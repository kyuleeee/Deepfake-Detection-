### GenConViT weight 다운로드


```python
from huggingface_hub import snapshot_download
import os

repo_id = "Deressa/GenConViT"
local_dir = "model/GenConViT"   # 원하는 경로

os.makedirs(local_dir, exist_ok=True)

# 특정 확장자만 받고 싶으면 allow_patterns에 패턴 추가
# (예: *.pt, *.pth, *.bin, *.safetensors, *.yaml, *.json 등)
downloaded_path = snapshot_download(
    repo_id=repo_id,
    local_dir=local_dir,
    local_dir_use_symlinks=False,          # symlink 대신 실제 파일 복사
    # allow_patterns=["*.pt", "*.pth", "*.bin", "*.safetensors", "*.yaml", "*.json"]
)

print("Saved to:", downloaded_path)
```

### Download Backbone CNN(ConvNeXt) weight


```python
import timm
import torch

bb_name = "convnext_tiny"
bb_path = "model/convnext_tiny.pth"

bb = timm.create_model(bb_name, pretrained=True)
torch.save(bb.state_dict(), bb_path)
```

### Download Embedding(Swin Transformer) ViT weight


```python
em_name = "swin_tiny_patch4_window7_224"
em_path = "model/swin_tiny_patch4_window7_224.pth"

em = timm.create_model(em_name, pretrained=True)
torch.save(em.state_dict(), em_path)
```
